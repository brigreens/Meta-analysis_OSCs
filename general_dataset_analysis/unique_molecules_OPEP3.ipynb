## Let's examine how many unique D/A pairs and unique molecules we have in the dataset

```python
from rdkit import Chem
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Open the dataset csv

```python
df = pd.read_csv('../OSC_dataset_metaanalysis.csv')
df
```

How many unique NFAs do we have (with and without side chains)?

```python
unique_smiles = []

for x in range(len(df['acc_SMILES_sidechains'])):
    try:
        # convert each to canonical smiles for fair comparison
        canon_smi = Chem.MolToSmiles(Chem.MolFromSmiles(df['acc_SMILES_sidechains'][x]))

        if canon_smi not in unique_smiles:
            unique_smiles.append(canon_smi)
    except:
        print(x)
        print(df['acc_SMILES_sidechains'][x])

print('The number of unique NFAs with side chains is: %s' % len(unique_smiles))
```

```python
unique_smiles = []

for x in range(len(df['acc_SMILES_shortened_sidechains'])):
    try:
        # convert each to canonical smiles for fair comparison
        canon_smi = Chem.MolToSmiles(Chem.MolFromSmiles(df['acc_SMILES_shortened_sidechains'][x]))

        if canon_smi not in unique_smiles:
            unique_smiles.append(canon_smi)
    except:
        print(x)
        print(df['acc_SMILES_shortened_sidechains'][x])

print('The number of unique NFAs with shortened side chains is: %s' % len(unique_smiles))
```

```python
unique_pairs = []
unique_smiles_NFA = []
unique_smiles_donors = []

for x in range(len(df['don_SMILES_sidechains'])):
    try:
        # convert each to canonical smiles for fair comparison
        canon_smi_don = Chem.MolToSmiles(Chem.MolFromSmiles(df['don_SMILES_sidechains'][x]))
        canon_smi_acc = Chem.MolToSmiles(Chem.MolFromSmiles(df['acc_SMILES_sidechains'][x]))

        canon_pair = [canon_smi_don, canon_smi_acc]

        if canon_pair not in unique_pairs:
            unique_pairs.append(canon_pair)
    except:
        print(x)

print('The number of unique pairs with full side chains is: %s' % len(unique_pairs))
```

### Lets save these as filtered dataset csvs. With sidechains means it was filtered by removing duplicates among SMILES that contain the sidechains. No sidechains means only the SMILES with shortened sidechains were looked at.

The csv was manually ordered by PCE, so highest PCE is first and lowest PCE is last. This is necessary so we can select the top performing pair for the filtered dataset

```python
df_filtered = pd.DataFrame(columns=['reported_acceptor','reported_donor', 'reference', 'DOI', 'Voc', 'Jsc',	'FF', 'PCE', 'acceptor_type', 'acceptor_class', 'fused_rings_in_core', 'donor_type', 'acc_SMILES_sidechains',	'acc_SMILES_shortened_sidechains', 'don_SMILES_sidechains',	'don_SMILES_shortened_sidechains'])

unique_pairs = []

for x in range(len(df['don_SMILES_sidechains'])):
    try:
        # convert each to canonical smiles for fair comparison
        canon_smi_don = Chem.MolToSmiles(Chem.MolFromSmiles(df['don_SMILES_sidechains'][x]))
        canon_smi_acc = Chem.MolToSmiles(Chem.MolFromSmiles(df['acc_SMILES_sidechains'][x]))

        canon_pair = [canon_smi_don, canon_smi_acc]

        if canon_pair not in unique_pairs:
            unique_pairs.append(canon_pair)
            df_filtered.loc[x] = df.iloc[x]
    except:
        pass

df_filtered.to_csv('OSC_filtered_with_sidechains.csv')
```
